```python
url = 'https://raw.githubusercontent.com/kalyaniuniversity/mgx-datasets/master/DLBCL/datasets/dlbcl-fl.csv'
```

# Importing Pandas

```python
import pandas as pd
```

# Reading CSV

```python
def readCSV(url):
    df = pd.read_csv(url)
    return df
```

# Printing the data

```python
readCSV(url)
```

# Counting the number of rows in CSV

```python
def row_count(url):
    return readCSV(url).shape[0] + 1

row_count(url)
```

# Counting the number of columns

```python
def column_count(url):
    return readCSV(url).shape[1]

column_count(url)
```

# Printing the first row of CSV

```python
def print_first_row_of_csv(url):
    df = pd.read_csv(url, skipfooter = len(readCSV(url).index), header = None, engine = 'python')
    return df
    
print_first_row_of_csv(url)
```

# Printing the CSV without the first row

```python
def print_without_1_row(url):
    return pd.read_csv(url, skiprows = 1, header = None)

print_without_1_row(url)
```

# Printing the dataset without last column of CSV

```python
def print_without_last_col(url):
    df = readCSV(url)
    df = df.drop(['class'],axis = 1)
    return df

print_without_last_col(url)
```

# Printing the last column of dataset

```python
def print_last_column(url):
    df = pd.read_csv(url)
    return df.drop(df.columns[:-1], axis = 1)

print_last_column(url)
```

# Printing the data without column Labels and target class

```python
def only_data(url):
    df = pd.read_csv(url, skiprows = 1, header = None)
    df = df.drop(df.columns[-1], axis = 1)
    return df

only_data(url)
```

# Printing Unique class labels

```python
def unique_class_labels(url):
    df = pd.read_csv(url)
    df = df.drop(df.columns[:-1], axis = 1)
    return df.drop_duplicates()
    
unique_class_labels(url)
```
